# Bansal-Yaron model

## Specification

The law of motion of state variables $(x_t,\sigma_t^2)$ and consumption and dividend growth

$$
\begin{align}
    x_{t+1}&=\rho x_t+\varphi_e\sigma_t{\sf q}e_{t+1}\\
    \sigma^2_{t+1}&=\sigma^2+\nu(\sigma^2_t-\sigma^2)+\sigma_w{\sf q}w_{t+1}\\
    c_{t+1}-c_t&=\mu_c+x_t+\sigma_t{\sf q}\eta_{t+1}\\
    d_{t+1}-d_t&=\mu_d+\phi x_t+\varphi_d\sigma_t{\sf q}u_{t+1}
\end{align}
$$

Epstein-Zin preference

$$
\begin{align}
    V_t&=((1-\beta)(C_t)^{1-\rho}+\beta(R_t)^{1-\rho})^{\frac{1}{1-\rho}}\\
    R_t&=\mathbb{E}[(V_{t+1})^{1-\gamma}]^{\frac{1}{1-\gamma}}
\end{align}
$$

In logarithmic form

$$
\begin{align}
    v_t-c_t&=\frac{1}{1-\rho}\log(1-\beta+\beta(r_t-c_t)^{1-\rho})\\
    r_t-c_t&=\frac{1}{1-\gamma}\log\mathbb{E}[\exp(1-\gamma)(v_{t+1}-c_t)]
\end{align}
$$

**LPH proposes to model the law of motion of the standard deviation $Y_{t+1}$ instead of variance $\sigma^2_{t+1}$, and also normalize the mean of $Y_{t+1}$ to 1 by scaling $\sigma_y,\sigma_z,\sigma_c,\sigma_d$ by $1/\sigma$. The results may change because variance as a state variable is replaced by standard deviation. To facilitate comparison with Bansal Yaron (2004), we follow the model in the paper in the below calculations.**

$$
\begin{align}
    Z_{t+1}&=\alpha_zZ_t+Y_t\sigma_z{\sf q}e_{t+1}\\
    Y_{t+1}&=1+\alpha_y(Y_t-1)+\sigma_y{\sf q}w_{t+1}\\
    c_{t+1}-c_t&=\eta_c+Z_t+Y_t\sigma_c{\sf q}\eta_{t+1}\\
    d_{t+1}-d_t&=\eta_d+Z_t+Y_t\sigma_d{\sf q}u_{t+1}
\end{align}
$$

```python
# define model dimension / number of variables

nY = 2 # number of dependent variables (g_c, g_d)
nX = 2 # number of state variables (x, sigma_sq)
nW = 4 # number of shocks
shape = (nY, nX, nW)
```

```python
# define model parameters

param_EZ = {
    'beta': .998,
    'rho': 1/1.5,
    'gamma': 10}

param_BY = {
    'rho': 0.979,
    'phi_e': 0.044,
    'sigma': 0.0078,
    'nu': 0.987,
    'sigma_w': 0.23e-5,
    'mu_c': 0.0015,
    'mu_d': 0.0015,
    'phi': 3,
    'phi_d': 4.5}
```

```python
# define model equilibrium conditions

def lm(X, W, q, param_EZ=param_EZ, param_BY=param_BY):
    x, s = X.ravel()
    sigma = anp.sqrt(s)
    e, w, eta, u = W.ravel()
    x_next = param_BY['rho'] * x + param_BY['phi_e'] * sigma * e
    s_next = anp.square(param_BY['sigma']) + param_BY['nu'] * (s - anp.square(param_BY['sigma'])) + param_BY['sigma_w'] * w
    return anp.array([[x_next], [s_next]])

def dv(X, W, q, param_EZ=param_EZ, param_BY=param_BY):
    x, s = X.ravel()
    sigma = anp.sqrt(s)
    e, w, eta, u = W.ravel()
    g_c_next = param_BY['mu_c'] + x + sigma * eta
    g_d_next = param_BY['mu_d'] + param_BY['phi'] * x + param_BY['phi_d'] * sigma * u
    return anp.array([[g_c_next], [g_d_next]])

def ge(X, X_next, W, q, param_EZ=param_EZ, param_BY=param_BY):
    vmc, rmc, x, s = X.ravel()
    sigma = anp.sqrt(s)
    vmc_next, rmc_next, x_next, s_next = X_next.ravel()
    sigma_next = anp.sqrt(s_next)
    e, w, eta, u = W.ravel()
    
    x_next_hat = param_BY['rho'] * x + param_BY['phi_e'] * sigma * e
    s_next_hat = anp.square(param_BY['sigma']) + param_BY['nu'] * (s - anp.square(param_BY['sigma'])) + param_BY['sigma_w'] * w

    g_c_next = param_BY['mu_c'] + x + sigma * eta
    g_d_next = param_BY['mu_d'] + param_BY['phi'] * x + param_BY['phi_d'] * sigma * u

    ell = vmc_next + g_c_next - rmc
    eq1 = (1-param_EZ['beta']) + param_EZ['beta'] * anp.exp((1-param_EZ['rho'])*rmc) - anp.exp((1-param_EZ['rho'])*vmc)
    eq2 = x_next - x_next_hat
    eq3 = s_next - s_next_hat
    
    return anp.array([ell, eq1, eq2, eq3])

def ss(param_EZ=param_EZ, param_BY=param_BY):
    vmc = (np.log(1-param_EZ['beta']) - np.log(1-param_EZ['beta']*np.exp((1-param_EZ['rho'])*param_BY['mu_c']))) / (1-param_EZ['rho'])
    rmc = vmc + param_BY['mu_c']
    x = 0
    s = param_BY['sigma']**2
    return anp.array([vmc, rmc, x, s])
```

```python
# define variable names for displaying LinQuad in latex

xName = [r'x', r'\sigma^2']
wName = [r'w_{}'.format(i+1) for i in range(4)]

def yName(yLabl):
    
    lst = {
        'g_c': r'c_{t+1}-c_t',
        'g_d': r'd_{t+1}-d_t',
        'vmc': r'v_t-c_t',
        'rmc': r'r_t-c_t',
        'vmr': r'v_{t+1}-r_t',
        'sdf': r's_{t+1}-s_t',
        'llh': r'\log N^*_{t+1}',
        'PD': r'\mathrm{eigenfunction}',
        'WC': r'\mathrm{eigenfunction}'}
    for name in ['g_c', 'g_d', 'vmc', 'rmc']:
        for i in range(3):
            lst['{}_{}'.format(name, i)] = lst[name].replace('_', '^{}_'.format(i))

    try:
        if '_' in yLabl and len(yLabl.rsplit('_', 1)[1]) > 5:
            yLabl, yMode = yLabl.rsplit('_', 1)
            return r'(##labl##)^{\mathrm{##mode##}}'.replace('##labl##', lst[yLabl]).replace('##mode##', yMode)
        else:
            return lst[yLabl]
    except: return '(unnamed)'

pName = ['Protocol 1', 'Protocol 2', 'Expansion Suite']
mName = ['expansion', 'expectation', 'hybrid']
```

## Solution: Exogenous consumption

1. Solve for second order expansion of $c_{t+1}-c_t$ and $v_t-c_t$ and $r_t-c_t$ (dependent on protocol)

2. Approximate, for ${\sf q}=1$ such that $\gamma=\gamma_0$ for both protocol,

$$
\begin{eqnarray}
    &&c_{t+1}-c_t &\approx& g_c+{\sf q}(c^1_{t+1}-c^1_t)+\frac{1}{2}{\sf q}^2(c^2_{t+1}-c^2_t)^2 \\
    &&v_t-c_t     &\approx& g_{v-c}+{\sf q}(v^1_t-c^1_t)+\frac{1}{2}{\sf q}^2(v^2_{t+1}-c^2_t)^2
\end{eqnarray}
$$

3. Use law of motion of state variables to compute $v_{t+1}-c_{t+1}$.

4. Compute, for ${\sf q}=1$,

$$
\begin{eqnarray}
    &&(r_t-c_t)^{\text{expansion}}   &\approx& g_{r-c}+{\sf q}(r^1_t-c^1_t)+\frac{1}{2}{\sf q}^2(r^2_{t+1}-c^2_t)^2 \\
    &&(r_t-c_t)^{\text{expectation}} &\approx& \frac{1}{1-\gamma}\log\mathbb{E}\left[\exp((1-\gamma)((v_{t+1}-c_{t+1})+(c_{t+1}-c_t)))\right]
\end{eqnarray}
$$

and the corresponding

$$
\begin{eqnarray}
    &&(v_{t+1}-r_t)^{\text{expansion}}   &\approx& (v_{t+1}-c_{t+1})+(c_{t+1}-c_t)-(r_t-c_t)^{\text{expansion}} \\
    &&(v_{t+1}-r_t)^{\text{expectation}} &\approx& (v_{t+1}-c_{t+1})+(c_{t+1}-c_t)-(r_t-c_t)^{\text{expectation}}
\end{eqnarray}
$$

5. Compute SDF

$$
\begin{eqnarray}
    &&(s_{t+1}-s_t)^{\text{expansion}}   &\approx& \log\beta-\rho(c_{t+1}-c_t)+(\rho-\gamma)(v_{t+1}-r_t)^{\text{expansion}} \\
    &&(s_{t+1}-s_t)^{\text{expectation}} &\approx& \log\beta-\rho(c_{t+1}-c_t)+(\rho-\gamma)(v_{t+1}-r_t)^{\text{expectation}} \\
    &&(s_{t+1}-s_t)^{\text{hybrid}}      &\approx& \log\beta-\rho(c_{t+1}-c_t)+(\rho-1)(v_{t+1}-r_t)^{\text{expansion}}+\underbrace{(1-\gamma)(v_{t+1}-r_t)^{\text{expectation}}}_{\log N^*_{t+1}}
\end{eqnarray}
$$

6. Compute wealth-consumption (and similarly, price-dividend) ratio

$$
\mathbb{Q}1(x)=\sum_{\tau=0}^{\infty}\mathbb{E}\left[\frac{S_{t+\tau}}{S_t}\frac{C_{t+\tau}}{C_t}\mid X_t=x\right]
$$

where $\log\frac{S_{t+1}}{S_t}=s_{t+1}-s_t$ and $\log\frac{C_{t+1}}{C_t}=c_{t+1}-c_t$ and $X_t$ is state variable.

Wealth-consumption ratio is also given by
$$
    \frac{W_t}{C_t}=\frac{1}{1-\beta}\left(\frac{V_t}{C_t}\right)^{1-\rho}.
$$

```python
# compute mean and std dev of state variables

mux = 0
sdx = param_BY['phi_e']*param_BY['sigma']/np.sqrt(1-param_BY['rho']**2)
muy = param_BY['sigma']**2
sdy = param_BY['sigma_w']/np.sqrt(1-param_BY['nu']**2)
```

```python
# compute steady state

x = lm_fp(lm, nX, nW)
w = np.zeros((nW, 1))
q = 0

# compute law of motion of model variables

dlm = lm_deri(lm, (x, w, q))

X0_next = LinQuad(
    coefs = {'c': x},
    shape = (nX, nX, nW))

X1_next = LinQuad(
    coefs = {
        'c': dlm['q'],
        'x': dlm['x'],
        'w': dlm['w']},
    shape = (nX, nX, nW))

X2_next = LinQuad(
    coefs = {
        'c': dlm['qq'],
        'x': dlm['xq']*2,
        'w': dlm['wq']*2,
        'x2': dlm['x'],
        'xx': dlm['xx'],
        'xw': dlm['xw']*2,
        'ww': dlm['ww']},
    shape = (nX, nX, nW))

# compute law of motion of dependent variables

ddv = lm_deri(dv, (x, w, q))

Y0 = LinQuad(
    coefs = {'c': ddv[' ']},
    shape = (nY, nX, nW))

Y1 = LinQuad(
    coefs = {
        'c': ddv['q'],
        'x': ddv['x'],
        'w': ddv['w']},
    shape = (nY, nX, nW))

Y2 = LinQuad(
    coefs = {
        'c': ddv['qq'],
        'x': ddv['xq']*2,
        'w': ddv['wq']*2,
        'x2': ddv['x'],
        'xx': ddv['xx'],
        'xw': ddv['xw']*2,
        'ww': ddv['ww']},
    shape = (nY, nX, nW))

# compute expansion of exogenous growth processes

g_c_0, g_d_0 = split(Y0)
g_c_1, g_d_1 = split(Y1)
g_c_2, g_d_2 = split(Y2)

gc0 = float(g_c_0['c'])

# compute vmc_0 and rmc_0

vmc_0 = LinQuad(
    coefs = {'c': 1/(1-param_EZ['rho']) * np.log((1-param_EZ['beta']) / (1-param_EZ['beta']*np.exp((1-param_EZ['rho'])*gc0)))},
    shape = (1,nX,nW))
rmc_0 = lag(vmc_0) + g_c_0

# compute lambda for computing higher order terms

lmbda = param_EZ['beta'] * np.exp((1-param_EZ['rho'])*gc0)
```

### Protocol 1

```python
Ew = np.zeros((nW, 1))
Eww = E_ww(Ew, np.eye(nW))

# compute vmc_1

cons = lmbda * E(g_c_1, Ew, Eww)
vmc_1 = vmc_solver(X1_next, X2_next, cons, lmbda, Ew, Eww, False)

# compute vmc_2

temp1a = lag(vmc_1, X1_next) + g_c_1
temp1b = E(temp1a.kron_prod(temp1a), Ew, Eww)
temp1c = E(temp1a, Ew, Eww)
temp1d = temp1c.kron_prod(temp1c)
temp1 = E(g_c_2, Ew, Eww) + (1-param_EZ['gamma']) * (temp1b - temp1d)
temp2 = (1-param_EZ['rho'])*(1-lmbda)/lmbda * vmc_1.kron_prod(vmc_1)
cons = lmbda * temp1 + temp2
vmc_2 = vmc_solver(X1_next, X2_next, cons, lmbda, Ew, Eww, True)

# compute rmc_1 and rmc_2

rmc_1 = temp1c
rmc_2 = E(lag(vmc_2, X1_next, X2_next), Ew, Eww) + temp1

p = {
    'X0_next': X0_next,
    'X1_next': X1_next,
    'X2_next': X2_next,
    'g_c_0': g_c_0, 'g_d_0': g_d_0,
    'g_c_1': g_c_1, 'g_d_1': g_d_1,
    'g_c_2': g_c_2, 'g_d_2': g_d_2,
    'vmc_0': vmc_0, 'rmc_0': rmc_0,
    'vmc_1': vmc_1, 'rmc_1': rmc_1,
    'vmc_2': vmc_2, 'rmc_2': rmc_2}
```

```python
q = 1
gamma = param_EZ['gamma']

# evaluate everything at q

g_c = approx(q, p['g_c_0'], p['g_c_1'], p['g_c_2'])
g_d = approx(q, p['g_d_0'], p['g_d_1'], p['g_d_2'])
vmc = approx(q, p['vmc_0'], p['vmc_1'], p['vmc_2'])

p['g_c'] = g_c
p['g_d'] = g_d
p['vmc'] = vmc

# compute sdf using expectation

rmc_expectation = log_E_exp((1-gamma) * (lag(vmc, X1_next, X2_next) + g_c)) / (1-gamma)
vmr_expectation = lag(vmc, X1_next, X2_next) + g_c - rmc_expectation
sdf_expectation = np.log(param_EZ['beta']) - param_EZ['rho']*g_c + (param_EZ['rho']-gamma) * vmr_expectation
llh_expectation = (1-gamma) * vmr_expectation
WC_expectation = Q_mapping(sdf_expectation + g_c, LinQuad({}, (1,nX,nW)), X1_next, X2_next, ())
PD_expectation = Q_mapping(sdf_expectation + g_d, LinQuad({}, (1,nX,nW)), X1_next, X2_next, ())

p['rmc_expectation'] = rmc_expectation
p['vmr_expectation'] = vmr_expectation
p['sdf_expectation'] = sdf_expectation
p['llh_expectation'] = llh_expectation
p['WC_expectation'] = WC_expectation
p['PD_expectation'] = PD_expectation

# compute sdf using expansion

rmc_expansion = approx(q, p['rmc_0'], p['rmc_1'], p['rmc_2'])
vmr_expansion = lag(vmc, X1_next, X2_next) + g_c - rmc_expansion
sdf_expansion = np.log(param_EZ['beta']) - param_EZ['rho']*g_c + (param_EZ['rho']-gamma) * vmr_expansion
llh_expansion = (1-gamma) * vmr_expansion
WC_expansion = Q_mapping(sdf_expansion + g_c, LinQuad({}, (1,nX,nW)), X1_next, X2_next, ())
PD_expansion = Q_mapping(sdf_expansion + g_d, LinQuad({}, (1,nX,nW)), X1_next, X2_next, ())

p['rmc_expansion'] = rmc_expansion
p['vmr_expansion'] = vmr_expansion
p['sdf_expansion'] = sdf_expansion
p['llh_expansion'] = llh_expansion
p['WC_expansion'] = WC_expansion
p['PD_expansion'] = PD_expansion

# compute sdf using hybrid

sdf_hybrid = np.log(param_EZ['beta']) - param_EZ['rho']*g_c + (param_EZ['rho']-1) * vmr_expansion + llh_expectation
WC_hybrid = Q_mapping(sdf_hybrid + g_c, LinQuad({}, (1,nX,nW)), X1_next, X2_next, ())
PD_hybrid = Q_mapping(sdf_hybrid + g_d, LinQuad({}, (1,nX,nW)), X1_next, X2_next, ())

p['sdf_hybrid'] = sdf_hybrid
p['WC_hybrid'] = WC_hybrid
p['PD_hybrid'] = PD_hybrid
```

## Solution: Expansion Suite

Since Expansion Suite computes $\log N^*_{t+1}$ internally, in addition to `expansion`, `expectation`, and `hybrid`, we also have

$$
(s_{t+1}-s_t)^{\text{internal}}=\log\beta-\rho(c_{t+1}-c_t)+(\rho-1)(v_{t+1}-r_t)^{\text{expansion}}+\log N^*_{t+1}
$$

**Please ignore `internal`**. It is supposed to be the same as `hybrid` but not due to an implementation error in Expansion Suite. It will be removed once the bug is fixed and it is confirmed to be equivalent to `hybrid`.

```python
nY = 2
nZ = 2
nW = 4
shape = (nY, nZ, nW)

ge_es = lambda X, X_next, W, q: ge(X, X_next, W, q, param_EZ, param_BY)
ss_es = lambda : ss(param_EZ, param_BY)

res = recursive_expansion(ge_es, ss_es, param_EZ, param_BY, shape)

def g(X, X_next, W, q, param_EZ=param_EZ, param_BY=param_BY):
    vmc, rmc, x, s = X.ravel()
    sigma = anp.sqrt(s)
    vmc_next, rmc_next, x_next, s_next = X_next.ravel()
    sigma_next = anp.sqrt(s_next)
    e, w, eta, u = W.ravel()
    g_c_next = param_BY['mu_c'] + x + sigma * eta
    g_d_next = param_BY['mu_d'] + param_BY['phi'] * x + param_BY['phi_d'] * sigma * u
    sdf_ex = anp.log(param_EZ['beta']) + (param_EZ['rho']-1)*(vmc_next + g_c_next - rmc) - param_EZ['rho']*g_c_next
    return anp.array([[g_c_next], [g_d_next], [sdf_ex]])

vmc_0, rmc_0 = split(LinQuad({'c': res.ss[:2]}, (2,nX,nW)))
vmc_1, rmc_1 = split(res.X1)[:2]
vmc_2, rmc_2 = split(res.X2)[:2]

p = {
    'X0_next': LinQuad({'c': res.ss[nY:].reshape((nZ,1))}, (nZ, nZ, nW)),
    'X1_next': res.Z1_next,
    'X2_next': res.Z2_next,
    'vmc_0': vmc_0, 'rmc_0': rmc_0,
    'vmc_1': vmc_1, 'rmc_1': rmc_1,
    'vmc_2': vmc_2, 'rmc_2': rmc_2}
```

```python
q = 1
gamma = 1 - (1-param_EZ['gamma']) / q

# evaluate everything at q

g_c = res.approximate_fun(lambda X, X_next, W, q: g(X, X_next, W, q)[0][0])
g_d = res.approximate_fun(lambda X, X_next, W, q: g(X, X_next, W, q)[1][0])
vmc = approx(q, p['vmc_0'], p['vmc_1'], p['vmc_2'])

p['g_c'] = g_c
p['g_d'] = g_d
p['vmc'] = vmc

# compute sdf using expectation

rmc_expectation = log_E_exp((1-gamma) * (lag(vmc, res.Z1_next, res.Z2_next) + g_c)) / (1-gamma)
vmr_expectation = lag(vmc, res.Z1_next, res.Z2_next) + g_c - rmc_expectation
sdf_expectation = np.log(param_EZ['beta']) - param_EZ['rho']*g_c + (param_EZ['rho']-gamma) * vmr_expectation
llh_expectation = (1-gamma) * vmr_expectation
WC_expectation = Q_mapping(sdf_expectation + g_c, LinQuad({}, (1,nX,nW)), res.Z1_next, res.Z2_next, ())
PD_expectation = Q_mapping(sdf_expectation + g_d, LinQuad({}, (1,nX,nW)), res.Z1_next, res.Z2_next, ())

p['rmc_expectation'] = rmc_expectation
p['vmr_expectation'] = vmr_expectation
p['sdf_expectation'] = sdf_expectation
p['llh_expectation'] = llh_expectation
p['WC_expectation'] = WC_expectation
p['PD_expectation'] = PD_expectation

# compute sdf using expansion

rmc_expansion = approx(q, p['rmc_0'], p['rmc_1'], p['rmc_2'])
vmr_expansion = lag(vmc, res.Z1_next, res.Z2_next) + g_c - rmc_expansion
sdf_expansion = np.log(param_EZ['beta']) - param_EZ['rho']*g_c + (param_EZ['rho']-gamma) * vmr_expansion
llh_expansion = (1-gamma) * vmr_expansion
WC_expansion = Q_mapping(sdf_expansion + g_c, LinQuad({}, (1,nX,nW)), res.Z1_next, res.Z2_next, ())
PD_expansion = Q_mapping(sdf_expansion + g_d, LinQuad({}, (1,nX,nW)), res.Z1_next, res.Z2_next, ())

p['rmc_expansion'] = rmc_expansion
p['vmr_expansion'] = vmr_expansion
p['sdf_expansion'] = sdf_expansion
p['llh_expansion'] = llh_expansion
p['WC_expansion'] = WC_expansion
p['PD_expansion'] = PD_expansion

# compute sdf using hybrid

sdf_hybrid = np.log(param_EZ['beta']) - param_EZ['rho']*g_c + (param_EZ['rho']-1) * vmr_expansion + llh_expectation
WC_hybrid = Q_mapping(sdf_hybrid + g_c, LinQuad({}, (1,nX,nW)), X1_next, X2_next, ())
PD_hybrid = Q_mapping(sdf_hybrid + g_d, LinQuad({}, (1,nX,nW)), X1_next, X2_next, ())

p['sdf_hybrid'] = sdf_hybrid
p['WC_hybrid'] = WC_hybrid
p['PD_hybrid'] = PD_hybrid

# compute sdf using internal

sdf = res.approximate_fun(lambda X, X_next, W, q: g(X, X_next, W, q)[2][0]) + res.log_M
WC = Q_mapping(sdf + g_c, LinQuad({}, (1,nX,nW)), res.Z1_next, res.Z2_next, ())
PD = Q_mapping(sdf + g_d, LinQuad({}, (1,nX,nW)), res.Z1_next, res.Z2_next, ())

p['sdf_internal'] = sdf
p['llh_internal'] = res.log_M
p['WC_internal'] = WC
p['PD_internal'] = PD
```

```python
# prepare data

def ztest(xyGrid, p):
    N = int(np.sqrt(xyGrid.shape[1]))
    z1Grid = xyGrid[0,:]**3 - xyGrid[0,:] + xyGrid[1,:]*100 + p['WC']*100
    z2Grid = xyGrid[0,:]**3 - xyGrid[0,:] + xyGrid[1,:]*100 + p['PD']*100
    return z1Grid.reshape((N,N)), z2Grid.reshape((N,N))

def zfunc(xyGrid, p):
    """
    z0Grid : W/C computed using V/C
    z1Grid : W/C computed using Q mapping
    z2grid : P/D computed using Q mapping
    z3grid : log(SDF)
    z4grid : log(E[N*])
    z5grid : W/C eigenvector
    z6grid : P/D eigenvector
    """
    N = int(np.sqrt(xyGrid.shape[1]))
    zGrid = {}
    zGrid['WC_vmc'] = np.exp(p['vmc'].multi_call(xyGrid, xyGrid) * (1-param_EZ['rho']) - np.log(1-param_EZ['beta']))
    zGrid['WC_qmap'] = Q_mapping_scalar_many(p['sdf'] + p['g_c'], LinQuad({}, (1,nX,nW)), p['X1_next'], p['X2_next'], xyGrid, xyGrid)
    zGrid['PD_qmap'] = Q_mapping_scalar_many(p['sdf'] + p['g_d'], LinQuad({}, (1,nX,nW)), p['X1_next'], p['X2_next'], xyGrid, xyGrid)
    zGrid['log_sdf'] = p['sdf'].multi_call(xyGrid, xyGrid)
    zGrid['lee_llh'] = log_E_exp(p['llh']).multi_call(xyGrid, xyGrid)
    zGrid['WC_eig'] = p['WC'][1].multi_call(xyGrid, xyGrid)
    zGrid['PD_eig'] = p['PD'][1].multi_call(xyGrid, xyGrid)
    for z in zGrid:
        zGrid[z] = zGrid[z].reshape((N,N))
    return zGrid

xGrid, yGrid = np.meshgrid(x, y)
xyGrid = np.concatenate((xGrid.reshape((N**2,-1)).T, yGrid.reshape((N**2,-1)).T))

z1Grid = zfunc(xyGrid, p1)
z2Grid = zfunc(xyGrid, p2)
z3Grid = zfunc(xyGrid, p3)
```
